```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
sh1 = 'All_times_data'
all_times = pd.read_excel('data/Consolidated Data.xlsx',sheet_name=sh1)
prefixes = [all_times.columns[1+4*i] for i in range(11)]
l = ['Entity']
for i in range(1,len(all_times.columns)):
    l.append(prefixes[(i-1)//4]+' '+all_times.loc[0,all_times.columns[i]])
all_times.columns = l
all_times.drop(0,axis=0,inplace=True)
all_times.set_index('Entity',inplace=True)
all_times.head()
```

```python
def convert_utilization(c,r,time_ref):
    if type(all_times[c][r]) != datetime:
        ref_date = 'Dec 31 1899'
        d1 = str(all_times[c][r])
        if d1.find('.') <0:
            d1 = d1+".00"
        d = ref_date+' '+d1
        test1 = datetime.strptime(d, '%b %d %Y %H:%M:%S.%f')-time_ref
    else:
        test1 = all_times[c][r]-time_ref
    minutes = test1.total_seconds()/60
    m1 = minutes%60
    hours = minutes//60
    return hours+m1/100.0

time_ref = datetime.strptime('Dec 31 1899  00:00:00.0', '%b %d %Y %H:%M:%S.%f')

for i in range(11):
    c = 3+4*i
    for j in all_times.index:
        all_times.loc[j,all_times.columns[c]] = convert_utilization(all_times.columns[c],j,time_ref)

all_times.head()
    
```

```python
def get_monthly_port_plots(port,col_sfx,ylabel,prefixes,data,t='line_plot',title='Some plot'):
    colors = ['green','blue','orange','black']
    port_lines = {'JNPT':['C/1','C/2','C/6','C/8'],
                  'NSICT': ['C/4','C/5'],
                  'GTI':['G/1','G/2','G/3'],
                  'BMCT':['BM/1','BM/2','BM/3','BM/4']}
    l = port_lines[port]
    plt.figure(figsize=(7,7))
    p = 0
    for j in l:
        if t=='line_plot':
            plt.plot(prefixes,[data[i+' '+col_sfx][j] for i in prefixes],color=colors[p],label=j)
            p += 1
    plt.legend()
    plt.title(title,size=17)
    plt.xlabel('Month',size=15)
    plt.ylabel(ylabel,size=15)
    plt.tick_params(which='both',axis='both',labelsize=13)
    plt.xticks(rotation=45)
    plt.show()
    return
```

```python
sfx = ['AVG loading','AVG release ']
time_ref = datetime.strptime('Dec 31 1899  00:00:00.0', '%b %d %Y %H:%M:%S.%f')
for i in prefixes:
    ref_date = 'Dec 31 1899'
    for j in all_times.index:
        col_name = i+' '+sfx[0]
        c,r = col_name,j
        d1 = str(all_times[c][r])
        if type(all_times[c][r]) != datetime:
            if d1.find('.') <0:
                d1 = d1+".00"
            d = ref_date+' '+d1
            test1 = datetime.strptime(d, '%b %d %Y %H:%M:%S.%f')-time_ref
        else:
            test1 = all_times[c][r]-time_ref
        all_times.loc[j,col_name] = test1/pd.Timedelta(hours=1)
        
        col_name = i+' '+sfx[1]
        c,r = col_name,j
        d1 = str(all_times[c][r])
        if type(all_times[c][r]) != datetime:
            if d1.find('.') <0:
                d1 = d1+".00"
            d = ref_date+' '+d1
            test1 = datetime.strptime(d, '%b %d %Y %H:%M:%S.%f')-time_ref
        else:
            test1 = all_times[c][r]-time_ref
        all_times.loc[j,col_name] = test1/pd.Timedelta(hours=1)
```

# PURE & MIX

```python
sh1 = 'Pure_Mix_data'
pure_mix = pd.read_excel('data/Consolidated Data.xlsx',sheet_name=sh1)
pure_mix.head()
cols = ['Port']
l = ['Rakes','Port','Rly']
l2 = ['Pure','Mix']
for i in prefixes:
    for j in l2:
        for k in l:
            cols.append(i+'_'+j+'_'+k)
pure_mix.columns = cols
pure_mix.drop([0,1],axis=0,inplace=True)
pure_mix.set_index('Port',inplace=True)
pure_mix.head(10)
```

```python
def convert_to_timedelta(data,c,r,time_ref):
    if type(data[c][r]) != datetime:
        ref_date = 'Dec 31 1899'
        d1 = str(data[c][r])
        if d1.find('.') <0:
            d1 = d1+".00"
        d = ref_date+' '+d1
        try:
            test1 = datetime.strptime(d, '%b %d %Y %H:%M:%S.%f')-time_ref
        except:
            d += ".00"
            test1 = datetime.strptime(d, '%b %d %Y %H:%M:%S.%f')-time_ref
    else:
        test1 = data[c][r]-time_ref
    return test1/pd.Timedelta(hours=1)

time_ref = datetime.strptime('Dec 31 1899  00:00:00.0', '%b %d %Y %H:%M:%S.%f')
l = ['Port','Rly']
l2 = ['Pure','Mix']
for m in prefixes:
    for n in l2:
        for i in l:
            col_name = m+"_"+n+"_"+i
            for j in pure_mix.index:
                pure_mix.loc[j,col_name] = convert_to_timedelta(pure_mix,col_name,j,time_ref)

pure_mix.head()
```

```python
pure_port_cols = [i+"_Pure_Port" for i in prefixes]
pure_rly_cols = [i+"_Pure_Rly" for i in prefixes]
mix_port_cols = [i+"_Mix_Port" for i in prefixes]
mix_rly_cols = [i+"_Mix_Rly" for i in prefixes]
pure_rakes_cols = [i+"_Pure_Rakes" for i in prefixes]
mix_rakes_cols = [i+"_Mix_Rakes" for i in prefixes]
```

```python
ports = list(pure_mix.index)
port_pure_stats = {}
for i in ports:
    port_pure_stats[i] = [np.mean(pure_mix.loc[i,pure_port_cols]),np.std(pure_mix.loc[i,pure_port_cols])]

port_mix_stats = {}
for i in ports:
    port_mix_stats[i] = [np.mean(pure_mix.loc[i,mix_port_cols]),np.std(pure_mix.loc[i,mix_port_cols])]

for i in ports:
    print(i, port_pure_stats[i],port_mix_stats[i])
```

```python
rly_pure_stats = {}
for i in ports:
    rly_pure_stats[i] = [np.mean(pure_mix.loc[i,pure_rly_cols]),np.std(pure_mix.loc[i,pure_rly_cols])]

rly_mix_stats = {}
for i in ports:
    rly_mix_stats[i] = [np.mean(pure_mix.loc[i,mix_rly_cols]),np.std(pure_mix.loc[i,mix_rly_cols])]

for i in ports:
    print(i, rly_pure_stats[i],rly_mix_stats[i])
```

```python
plt.figure(figsize=(7,7))
plt.plot(prefixes,pure_mix.loc['JNPT',pure_port_cols],color='blue',label='Pure')
plt.plot(prefixes,pure_mix.loc['JNPT',mix_port_cols],color='red',label='Mix')
plt.legend()
plt.title('Pure vs Mix detention times (JNPT)',size=17)
plt.xlabel('Month',size=15)
plt.ylabel('Detention(h)',size=15)
plt.tick_params(which='both',axis='both',labelsize=13)
plt.xticks(rotation=45)
plt.show()
```
